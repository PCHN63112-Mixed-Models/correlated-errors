# Introduction
In Psychological and Cognitive Neuroscientific research, taking multiple measurements from a single subject is a common practice. Doing so had many advantages, from both a practical and statistical perspective. Not only do we make the most of single individual in terms of their time and effort, we also get statistical benefits in terms of increased power when using a smaller sample. Conceptually, we can also use each subject as their own control, allowing for an easy solution to problems of individual differences between experimental groups. Indeed, in many areas of behavioural research, these forms of designs are the *rule* rather than the exception. Unfortunately, such a seemingly small tweak in the design has rather larger ramifications when it comes to the analysis. Although traditional statistical education in psychology treats the analysis of these types of design as equal to the methods we discussed last semester, the reality is that taking multiple measurements from the *same* individual causes a mountain of additional complications.  

... In this lesson, we will examine the theory behind *why* linear models with correlated errors are problematic. We will then discuss traditional methods for dealing with correlated errors, within the context of repeated-measurement designs. Finally, we will briefly introduce linear mixed-effects models as the most modern way of dealing with these types of situations.
