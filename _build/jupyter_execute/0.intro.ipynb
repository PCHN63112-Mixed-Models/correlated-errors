# Introduction
In both psychological and cognitive neuroscientific research, taking multiple measurements from a single subject is a common practice. This has many advantages, from both a practical and statistical perspective. Not only do we make the most of a single individual in terms of their time and effort, we also get statistical benefits in terms of increased power when using a smaller sample. Conceptually, we can also use each subject as their own control, allowing for an easy solution to problems of individual differences between experimental groups. Indeed, in many areas of behavioural research, these forms of designs are the *rule* rather than the exception. Unfortunately, such a seemingly small tweak in the design has rather larger ramifications when it comes to the analysis. Although traditional statistical education in psychology treats the analysis of these designs as equivalent to the methods we discussed last semester, the reality is that taking multiple measurements from the *same* individual causes many additional complications. 

In this lesson, we will examine the theory behind *why* repeated measurements cause problems. We will cover the theory behind this and will introduce a change to our probabilistic framework that allows these difficulties to be accommodated. We will then discuss traditional methods for dealing with repeated measures in terms of the paired $t$-test and the repeated measures ANOVA. As we will see, these traditional methods have a host of problems, both theoretical and practical. As such, the latter half of the lesson is really an exercise in motivating *why* we need a better approach. This will lay the ground-work for introducing mixed-effects models next week.
