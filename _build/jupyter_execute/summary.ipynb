# Summary
In this lesson we have discussed the nature of repeated measurement designs, the problems they cause to our analytic framework and one possible way of dealing with them, using the repeated measures ANOVA. In general, the core understanding to take away from this is that repeated measurements induce correlation in our data, which has a direct impact on the magnitude of the standard errors and thus directly informs the uncertainty we use for inference. The normal linear model assumes that the correlation between data points is 0 and thus will either *under-estimate* or *over-estimate* the true standard errors when applied to repeated measurements. To improve this, we have to move to a probabilistic framework that allows for correlation, which we have seen in the form of the multivariate normal distribution with a given variance-covariance structure. In an ideal world, we would want to simply estimate the covariance structure from the data and use that estimate to inform the standard errors. However, traditional methods of analysing repeated measurements do not do this. Either the correlation is effectively *removed* via subtraction, or the correlation structure is assumed to have a specific form. These traditional methods also attempt to use the normal linear model to analyse repeated measures data, which involves partitioning the single error term and then manually arranging the arithmetic of the ANOVA table. This is effectively an *ad hoc* arrangement used to analyse repeated measures within a framework that is not suitable. Although not too bad with simple data, as soon as more complex designs are introduced this becomes unwieldly. Because of this, the traditional repeated measures ANOVA has severe limitations both *theoretically* and *practically*. Although this was the only option in the past, more modern approaches have eclipsed the repeated measures ANOVA due to their ability to get round all these issues. These methods will be the focus of this part of the unit.
